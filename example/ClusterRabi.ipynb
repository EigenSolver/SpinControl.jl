```julia
using SpinEnsembles
import SpecialFunctions: gamma
```

Rabi Oscillation 
$$
    H=\beta S_z+h S_x,\quad h\gg|\beta|
$$

Analytical Solution in 3D

$$
M_z(t)=A \cos(h \,t+\varphi), \quad M_y(t)=-A \sin(h \,t+\varphi)
$$

$$
A=\frac{1}{2}\left(1+b^4t^2/h^2\right)^{-1/4},\quad \varphi=\frac{1}{2}\arctan(b^2 t/h)
$$

## $M_z(t)$

```julia
cluster=SpinCluster(SpinEnsemble(0.39486,3,[0,0,1],0.3,15,:spherical))
T2=coherencetime(cluster)
b=dipolarlinewidth(cluster)
```

```julia
h=20
t=collect(0:/(20*h):5*/h)

println("h=$h \n","b=$b")
println(h>10*b)
numerical_rabi=rabi(t,cluster,h, N=1000, axis=3)
theoretical_rabi=analyticalrabi(t,cluster,h,axis=3)
plot(t, theoretical_rabi, label="analytical",
xlabel="t",
ylabel="Sz(t)")
scatter!(t, numerical_rabi, color=4, label="numerical")
```

```julia
h_set=[50,100,200,400]

fig=plot(;xlabel="t",
ylabel="Sz(t)")

for h in h_set
    t=collect(0:(h/50)*300*/h:1200*T2)
    numerical_rabi=rabi(t,cluster,h, N=1500, axis=3)
    theoretical_rabi=analyticalrabi(t,cluster,h,axis=3)
    plot!(t, theoretical_rabi, color=h%9, label="analytical, h=$h")
    scatter!(t, numerical_rabi, color=h%9, label="numerical, h=$h",marker=:+, markersize=2)
end

display(fig)
```

```julia
h=100
t=collect(0:200/h:T2*400).+/h

println("h=$h \n","b=$b")
println(h>10*b)
numerical_rabi=rabi(t,cluster,h, N=500, axis=3)
theoretical_rabi=analyticalrabi(t,cluster,h,axis=3)
plot(t, theoretical_rabi, label="analytical",
xlabel="t",
ylabel="Sz(t)",
linestyle=:dash)
scatter!(t, numerical_rabi, color=4, label="numerical", marker=:+)
```
